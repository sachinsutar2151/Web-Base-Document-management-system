## Visualize quantitative data

### Subtask:
Generate plots for the 'Qty' column, such as histograms or box plots, to understand its distribution.


**Reasoning**:
Generate a histogram and a box plot for the 'Qty' column to visualize its distribution and identify outliers.



```python
plt.figure(figsize=(12, 6),facecolor='salmon')

plt.subplot(1, 2, 1)
sns.histplot(df1['Qty'], kde=True, bins=50)
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=df1['Qty'])
plt.title('Box Plot of Quantity')
plt.xlabel('Quantity')

plt.tight_layout()
plt.show()
```

## Visualize categorical data

### Subtask:
Generate plots for categorical columns like 'Item Name', 'Category', and 'Product Name' to show the frequency or distribution of values. This could involve bar plots or count plots.


**Reasoning**:
Create a figure and plot the count of each category.



```python
plt.figure(figsize=(10, 6),facecolor='lightcoral')
CM=sns.countplot(data=df1, x='Category',hue='Category',palette='plasma')
def data_label(arg):
  for i in arg.containers:
    arg.bar_label(i,fontsize=10,label_type="edge",color="black")
data_label(CM)
plt.title('Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

**Reasoning**:
Create a figure and plot the count of the top 10 product names to manage the number of categories displayed.



```python
plt.figure(figsize=(12, 6),facecolor='lightsteelblue')
top_products = df1['Product Name'].value_counts().nlargest(10).index
CM1=sns.countplot(data=df1[df1['Product Name'].isin(top_products)], x='Product Name',palette='rocket',order=top_products)
def data_label(arg):
  for i in arg.containers:
    arg.bar_label(i,fontsize=10,label_type="edge",color="black")
data_label(CM1)
plt.title('Distribution of Top 10 Product Names')
plt.xlabel('Product Name',color='black',fontsize=12)
plt.ylabel('Count',color='black',fontsize=12)
plt.xticks(color='navy')
plt.show()
```

**Reasoning**:
Create a figure and plot the count of the top 10 item names to manage the number of categories displayed.



```python
plt.figure(figsize=(12, 6),facecolor='black')
top_items = df1['Item Name'].value_counts().nlargest(10).index
sns.countplot(data=df1[df1['Item Name'].isin(top_items)], x='Item Name',hue='Item Name',palette='crest',order=top_items)
plt.title('Distribution of Top 10 Item Names',color='white')
plt.xlabel('Item Name',color='white',fontsize=13)
plt.ylabel('Count',color='white',fontsize=13)
plt.xticks(rotation=30, ha='right',color='white',fontsize=8)
plt.yticks(color='white',fontsize=8)
plt.show()
```

## Visualize time series data

### Subtask:
Create time-based visualizations using the 'Date', 'Month_Year', and 'Year' columns to observe trends or patterns in 'Qty' or other relevant columns over time. Line plots are suitable for this.


**Reasoning**:
Group the data by Date and sum the Quantity to prepare for daily trend visualization, then create a line plot for daily quantity.



```python
daily_qty = df1.groupby('Date')['Qty'].sum()
plt.figure(figsize=(15, 7),facecolor='navy')
daily_qty.plot(kind='line',grid=True)
plt.title('Daily Total Quantity Over Time',color='white',fontsize=13)
plt.xlabel('Date',color='white',fontsize=13)
plt.ylabel('Total Quantity',color='white',fontsize=13)
plt.xticks(rotation=30, ha='right',color='white',fontsize=8)
plt.yticks(color='white',fontsize=8)
plt.show()
```

**Reasoning**:
Group the data by Month_Year and sum the Quantity to prepare for monthly trend visualization, then create a line plot for monthly quantity and rotate x-axis labels for readability.



```python
monthly_qty = df1.groupby('Month_Year')['Qty'].sum()
plt.figure(figsize=(15, 7),facecolor='navy')
monthly_qty.plot(kind='line',grid=True)
plt.title('Monthly Total Quantity Over Time',color='white',fontsize=13)
plt.xlabel('Month Year',color='white',fontsize=13)
plt.ylabel('Total Quantity',color='white',fontsize=13)
plt.xticks(color='white')
plt.yticks(color='white')
plt.show()
```

**Reasoning**:
Group the data by Year and sum the Quantity to prepare for yearly trend visualization, then create a bar plot for yearly quantity.



```python
yearly_qty = df1.groupby('Year')['Qty'].sum()
plt.figure(figsize=(8, 8))
plt.pie(yearly_qty, labels=yearly_qty.index, autopct='%1.1f%%', startangle=0,explode=(0.1, 0, 0,0))
plt.title('Yearly Total Quantity Distribution')
plt.show()
```

## Explore relationships between columns

### Subtask:
Generate plots that show relationships between different columns, such as scatter plots for 'Date' and 'Qty', or bar plots showing 'Qty' by 'Category'.


**Reasoning**:
Create plots to visualize the relationships between 'Date' and 'Qty', 'Category' and 'Qty', and the top 10 'Product Name' and 'Item Name' by 'Qty'.



```python
plt.figure(figsize=(15, 7),facecolor='tomato')
sns.scatterplot(data=df1, x='Date', y='Qty')
plt.title('Quantity order by Date')
plt.xlabel('Date',fontsize=13)
plt.ylabel('Quantity',fontsize=13)
plt.show()

plt.figure(figsize=(12, 6),facecolor='tomato')
category_qty = df1.groupby('Category')['Qty'].sum().sort_values(ascending=False)
sns.barplot(x=category_qty.index, y=category_qty.values)
plt.title('Total Quantity by Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(12, 6),facecolor='tomato')
product_qty = df1.groupby('Product Name')['Qty'].sum().nlargest(10)
sns.barplot(x=product_qty.index, y=product_qty.values)
plt.title('Total Quantity by Top 10 Product Names')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(12, 6),facecolor='tomato')
item_qty = df1.groupby('Item Name')['Qty'].sum().nlargest(10)
sns.barplot(x=item_qty.index, y=item_qty.values)
plt.title('Total Quantity by Top 10 Item Names')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45, ha='right')
plt.show()
```
